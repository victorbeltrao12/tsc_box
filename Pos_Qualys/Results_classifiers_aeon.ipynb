```python
import pandas as pd
import numpy as np
import timeit
from datetime import timedelta
from aeon.benchmarking.results_loaders import get_estimator_results
from aeon.datasets.tsc_data_lists import univariate_equal_length
from aeon.registry import all_estimators
```

```python
all_estimators("classifier", filter_tags={"algorithm_type": "feature"})
```

```python
clfs_msc = ['Catch22Classifier','FreshPRINCEClassifier','TSFreshClassifier']
#SignatureClassifier, SummaryClassifier
```

```python
#Lista de classificadores
clfs = ['Arsenal', 'BOSS', 'CBOSS', 'CNN', 'CIF', 'Catch22', 'DrCIF',
        'FreshPRINCEClassifier', 'HIVECOTEV1', 'HIVECOTEV2', 'InceptionTimeClassifier', 'STC', 'TS-CHIEF',
        'ResNet', 'ProximityForest', 'RocketClassifier', 'TSFreshClassifier','WEASEL']



#todos os resultados dos classificadores
total_time = 0
start = timeit.default_timer()
all_results = get_estimator_results(clfs_msc, univariate_equal_length)
stop = timeit.default_timer()
total_time += stop - start
real_time = str(timedelta(seconds=total_time))
#Transformando os resultados em um dataframe
results_df = pd.DataFrame(all_results)

#Salvando os resultados em um parquet
results_df.to_csv('all_results_msc.csv', index=True)
print('Time runtime (seconds): ', real_time)
```
